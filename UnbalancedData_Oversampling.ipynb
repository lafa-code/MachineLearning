```python
import pandas as pd

file_path = '/content/drive/MyDrive/Internal-2024/Data_stroke.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
categorical_columns = ['gender', 'age', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
numeric_columns = ['hypertension', 'heart_disease', 'avg_glucose_level', 'bmi', 'stroke']

# ubah kolom kategori ke data tipe kategori
df[categorical_columns] = df[categorical_columns].astype('category')

# encoding
for col in categorical_columns:
    df[col] = df[col].cat.codes
```

```python
df.head()
```

```python
import pandas as pd
from imblearn.over_sampling import ADASYN, RandomOverSampler, SMOTE, SMOTEN, BorderlineSMOTE, KMeansSMOTE, SVMSMOTE
from sklearn.utils import shuffle

def balance_dataset(df, target_col, method_name):
    print("Class distribution sebelum balancing:")
    print(df[target_col].value_counts())

    X = df.drop(columns=[target_col])
    y = df[target_col]

    methods = {
        "ADASYN": ADASYN(),
        "RandomOversampler": RandomOverSampler(),
        "SMOTE": SMOTE(),
        "SMOTEN": SMOTEN(),
        "Borderline-SMOTE": BorderlineSMOTE(),
        "svm-SMOTE": SVMSMOTE()
    }
    if method_name not in methods:
        raise ValueError(f"Invalid method name: {method_name}. Choose from {list(methods.keys())}")

    sampler = methods[method_name]
    X_resampled, y_resampled = sampler.fit_resample(X, y)

    balanced_df = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=[target_col])], axis=1)
    balanced_df = shuffle(balanced_df).reset_index(drop=True)

    print("Class distribution sesudah balancing:")
    print(balanced_df[target_col].value_counts())

    save_path = f"/content/drive/MyDrive/01-DATASET_FINAL/Hasil/{method_name}_Data_Stroke_Balanced.csv"
    balanced_df.to_csv(save_path, index=False)
    print(f"Balanced dataset saved to: {save_path}")

    return balanced_df


""" Pilih salah satu metode:
        ADASYN
        RandomOversampler
        SMOTE
        SMOTEN
        Borderline-SMOTE
        svm-SMOTE
"""

method_name = "ADASYN"

```

```python
balance_df = balance_dataset(df, 'stroke', method_name)
```
